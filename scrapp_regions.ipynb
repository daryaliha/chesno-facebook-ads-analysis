```python
page_ids = list(set(map(lambda x:x["page_id"],adItems)))
```

```python
def disclaimer_iter(x):
    return x["funding_entity"] if "funding_entity" in x.keys() else ""
```

```python
def disclaimers_list_iter(page_id):
    pageAds = list(filter(lambda x: x["page_id"]==page_id,adItems))
    disclaimers_list = list(set(map( disclaimer_iter , pageAds)))
    clean_list = [x for x in disclaimers_list if x != '']
    return (",").join(clean_list)
```

```python
disclaimers = list(map(disclaimers_list_iter,page_ids))
```

```python
def name_iter(page_id):
    pageAds = list(filter(lambda x: x["page_id"]==page_id,adItems))

    return pageAds[0]["page_name"]
```

```python
pages_names = list(map(name_iter,page_ids))
```

```python
def date_iter(x):
#     print (month,x["ad_delivery_start_time"][:-3],x["ad_delivery_start_time"][:-3]==month)
    return x["ad_delivery_start_time"][:-3]==month
```

```python
def sum_lower_bound(x,y):
    if ("lower_bound" in y["spend"].keys()):
        return x+ int(y["spend"]["lower_bound"])
    else:
        return x
    

def sum_lower_upper(x,y):
    if ("upper_bound" in y["spend"].keys()):
        return x+ int(y["spend"]["upper_bound"])
    else:
        return x

```

```python
from functools import reduce
def dates_list_iter(page_id):
    pageAds = list(filter(lambda x: x["page_id"]==page_id,adItems))
    date_list = list(filter( date_iter , pageAds))
    lower_bound_sum = reduce(sum_lower_bound,date_list,0)
    return lower_bound_sum
def dates_list_iter_upper(page_id):
    pageAds = list(filter(lambda x: x["page_id"]==page_id,adItems))
    date_list = list(filter( date_iter , pageAds))
    lower_bound_sum = reduce(sum_lower_upper,date_list,0)
    return lower_bound_sum
```

```python
output_dictionary = {"Page Id": page_ids, "Disclaimer":disclaimers,"Page Name":pages_names}
```

```python
for month in months_list:
    dates_lower=list(map(dates_list_iter,page_ids))
    dates_upper=list(map(dates_list_iter_upper,page_ids))
    output_dictionary[month+"_lower"] = dates_lower   
    output_dictionary[month+"_upper"] = dates_upper
```
